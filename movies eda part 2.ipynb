```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
df2= pd.read_csv('movies/final_tmdb_data_2001.csv.gz')
df2
```

```python
# budget and revenue with values greater than 0
new_df = df2[(df2["budget"]>0) & (df2["revenue"]>0) ]
new_df.head()
```

```python
# MOVIES PER CERTIFICATION
new_df["certification"].value_counts()
```

```python
# BARPLOT

new_df["certification"].value_counts().plot.bar()
plt.title(' Movies per Certification', fontsize = 14,)
plt.xlabel('Certification', fontsize = 14)
plt.ylabel(' Movies', fontsize = 14);
plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8,);
plt.tight_layout()
plt.show;
```

```python
# movies per certification with missing values
new_df["certification"].isnull().sum()
```

```python
# ertification and extracting the average revenue per certification

revenue = new_df.groupby('certification')['revenue'].mean().round()
revenue
```

```python
revenue.plot.bar()
plt.ylabel("revenue")
plt.tight_layout()
plt.show;
```

```python
# certification and extracting the average budget per certification

budget = new_df.groupby('certification')['budget'].mean().round()
budget
```

```python
budget.plot.bar()
plt.ylabel("budget")
plt.tight_layout()
plt.show;
```
